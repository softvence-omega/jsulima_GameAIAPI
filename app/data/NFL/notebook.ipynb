```python
path = "nfl_games_data_history.csv"
df = pd.read_csv(path)
df.shape
```

```python
filtered_df = df[
    (df['home_team_name'] == 'Pittsburgh Steelers') & (df['away_team_name'] == 'Carolina Panthers') |
    (df['away_team_name'] == 'Pittsburgh Steelers') & (df['home_team_name'] == 'Carolina Panthers')
]
filtered_df.shape
```

```python
filtered_df[['home_team_name', 'away_team_name', 'home_total_score', 'away_total_score']]
```

```python
# Initialize win counters
steelers_wins = 0
panthers_wins = 0

# Iterate through each row
for _, row in filtered_df.iterrows():
    home_team = row['home_team_name']
    away_team = row['away_team_name']
    home_score = row['home_total_score']
    away_score = row['away_total_score']

    if home_score > away_score:
        winner = home_team
    elif away_score > home_score:
        winner = away_team
    else:
        winner = 'Draw'  # Optional, if you want to track ties

    if winner == 'Pittsburgh Steelers':
        steelers_wins += 1
    elif winner == 'Carolina Panthers':
        panthers_wins += 1

# Display results
print(f"Pittsburgh Steelers wins: {steelers_wins}")
print(f"Carolina Panthers wins: {panthers_wins}")
```
